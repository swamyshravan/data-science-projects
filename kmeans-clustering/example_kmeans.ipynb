The example is taken from "https://www.reneshbedre.com/blog/kmeans-clustering-python.html" for the purpose of learning kmeans clustering

```python
from sklearn.datasets import make_blobs
import pandas as pd
```

**Generating dataset**

```python
dataset, classes = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=0.5, random_state=0)
# make as panda dataframe for easy understanding
df = pd.DataFrame(dataset, columns=['var1', 'var2'])
df.head(2)
```

**Finding how many clusters are required**

```python
from yellowbrick.cluster import KElbowVisualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,12)).fit(df)
visualizer.show()
```

**Generating a kmeans model of 4 clusters**

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=0).fit(df)
```

**Examining the model we generated**

```python
kmeans.labels_  
#same as 
#kmeans.predict(df)
```

```python
kmeans.inertia_
```

```python
kmeans.n_iter_
```

```python
kmeans.cluster_centers_
```

**Getting each cluster size**

```python
from collections import Counter
Counter(kmeans.labels_)
#Counter({2: 50, 0: 50, 3: 50, 1: 50})
```

**Visualizing the cluster**

Visualising the cluster as scatter plot 

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(data=df, x="var1", y="var2", hue=kmeans.labels_)
plt.show()
```

Visualising the clusters with their centeroids

```python
sns.scatterplot(data=df, x="var1", y="var2", hue=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], 
            marker="x", c="r", s=80, label="centroids")
plt.legend()
plt.show()
```
